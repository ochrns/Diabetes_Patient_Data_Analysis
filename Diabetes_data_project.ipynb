```python
#count unique patient number

unique_patient_count = df['patient_nbr'].nunique()
print("Number of unique patient numbers:", unique_patient_count)
```

```python
#replacing ? with NaN in weight column
df['weight'] = df['weight'].replace('?', np.nan)
df.sample(40)
```

findings: Weight information is incomplete for large number of patients, making it unreliable for analysis. Due to extensive missing data, imputing values could introduce bias, so the column was dropped. This removes one feature but improves overall data integrity.

```python
#drop weight column

clean_df = df.drop('weight', axis=1)
clean_df
```

```python
#remove brackets in age column
clean_df["age"] = clean_df["age"].str.replace(r"[\[\]\(\)]", "", regex=True)
clean_df
```

The brackets in the age column were removed to ensure consistency.

```python
#group patients by gender and count the number of unique patient_nbr for male and female

gender_patient_count = clean_df.groupby('gender')['patient_nbr'].nunique()
gender_patient_count
```

The dataset contains more female patients (38026) than male patients (33492). There are 3 records with unknown gender classification, but this is not significant.

```python
medication_stats = clean_df.groupby('gender')['num_medications'].describe()
print(medication_stats)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='age', y='num_medications', hue=None, errorbar=None, legend=False, palette='viridis')
plt.title('Average Number of Medications by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Medications', fontsize=12)
plt.xticks(rotation=45)
plt.title('Average Number of Medications by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Medications', fontsize=12)
plt.show()
```

The average number of medications increases with age, peaking in the 60-70 age group. After there's a slight decline in the number of medications for older age groups.

```python
#pick unique patient_nbr and group patients by race, organize descending

patient_race_count = clean_df.groupby('race')['patient_nbr'].nunique().sort_values(ascending=False)
patient_race_count

```

Caucasian patients have highest count (53601), followed by African Americans (12932). Total of ? and Other gives us a number of 3186 patients with lost or unidentified race. The lowest number of Asian patients (517) may indicate they are less likely to suffer from diabetes.



```python
# Identify the patient with the highest number of visits
patient_visits = clean_df.groupby('patient_nbr')['encounter_id'].count()
patient_with_highest_visits = patient_visits.idxmax()
print(f"Patient with the highest visits: {patient_with_highest_visits}")

# display all columns for this patient
patient_data = clean_df[df['patient_nbr'] == patient_with_highest_visits]
print("\nData for the patient with the highest visits:")
patient_data
```

findings: the patient with the highest number of 40 visits is a Caucasian female in her 20s, who frequently arrives via Emergency department (admission_source_id=7). Her insulin dosage is often recorded as "Up" or "Down", indicating ongoing adjustments to manage her condition effectively.

```python
#time_in_hospital for patient 88785891

patient_data = clean_df[clean_df['patient_nbr'] == 88785891]
time_in_hospital_values = patient_data['time_in_hospital']

if not time_in_hospital_values.empty:
  print("Statistics for time_in_hospital for patient 88785891:")
  print(time_in_hospital_values.describe())
else:
  print("No data found for patient 88785891.")
```

```python
#select diag_1, diag_2, diag_3, for each visit for patient 88785891

patient_data = clean_df[clean_df['patient_nbr'] == 88785891]

most_frequent_diags = patient_data.groupby('encounter_id').agg(
    {
        'diag_1': lambda x: x.value_counts().idxmax(),
        'diag_2': lambda x: x.value_counts().idxmax(),
        'diag_3': lambda x: x.value_counts().idxmax(),
    }
)

most_frequent_diags
```

This patient has diabetes with ketoacidosis, hypothyroidism, and struggles with managing blood sugar levels, requiring regular medical supervision and medication. She also experience digestive issues, respiratory infections, and electrolyte imbalances, suggesting a complex health routine with dietary adjustments and hydration needs. Mental health challenges and frequent substance use may further impact their ability to maintain consistent self-care.

```python
readmission_percentages = clean_df["readmitted"].value_counts(normalize=True) * 100
print(round(readmission_percentages, 2))
```

This dataset shows 3 readmisson patterns:

*   No readmission: 53.91%
*   Readmission after 30 days (>30): 34.93%
*   Readmission within 30 days (<30): 11.16%



```python
# Visualize readmissions sorted by categorical values age, race, gender
for col in ["age", "race", "gender"]:
    plt.figure(figsize=(10, 4))
    order = clean_df[col].value_counts().index
    sns.countplot(data=clean_df, x=col, hue="readmitted", order=order, palette='viridis')
    plt.title(f"Readmission by {col}")
    plt.xticks(rotation=45)
    plt.legend(title="Readmitted")
    plt.show()
```

Patients in the age of 60-80 have the highest readmission rate.
Caucasians patients have higher readmission rate comparing to other race group.
Females patients have higher readmission rate than male patients.
